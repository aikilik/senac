```python
import matplotlib.pyplot as plt
import math

lista_tempo = []
lista_eixo_x = []
lista_eixo_y = []

x_0 = 0.0
y_0= 0.0
g = 9.8
v_0 = 30.0
angulo = 80 * math.pi / 180

v_0x = v_0 * math.cos(angulo)
v_0y = v_0 * math.sin(angulo)

tempo = 0.0
dt = 0.1

while tempo < 4.2:
    x_t = x_0 + v_0x * tempo
    y_t = y_0 + v_0y * tempo - 0.5 * g * tempo**2
    
    lista_tempo.append(tempo)
    lista_eixo_x.append(x_t)
    lista_eixo_y.append(y_t)
    
    tempo += dt

plt.xlabel('$distância$ ($m$)')
plt.ylabel('$altura$ (m)')
plt.plot(lista_eixo_x, lista_eixo_y)
plt.title('Movimento Projétil')

print(lista_tempo)
print(" ")
print(lista_eixo_x)
print(" ")
print(lista_eixo_y)

# y_t = y_0 + v_0y  *  tempo  -  0.5  *  g  *  tempo**2
# Não substitua  v_0y por v_y como fiz durante a aula.
```
