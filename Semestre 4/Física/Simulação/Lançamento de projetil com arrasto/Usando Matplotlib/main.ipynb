```python
import matplotlib.pyplot as plt
import math

lista_tempo = []
lista_eixo_x = []
lista_eixo_y = []

x_0 = 0.0
y_0= 0.0
g = 9.8
v_0 = 18.0
angulo = 80 * math.pi / 180

# dados para o fenomeno do arrasto
# A resistencia do ar é dada por r = k * (v**2)
# A constante k é dada por k = 1/2 * densidade * a_transversal * c_x 
# Assim, r = 1 / 2 * densidade * a_transveral * c_x
# p = m * g

densidade_ar = 1.225 # densidade do ar
a_transversal = 0.8 # é a maior área do corpo perpendicular à direção do movimento
c_x = 0.47 # coeficiente de arrasto aerodinâmico. Para uma bola o valor é de 0.5

k = 0.5 * densidade_ar * a_transversal * c_x # constante k que varia conforme o tipo de material utilizado
r = k * math.pow(v_0, 2)

massa = 0.0 # 
tempo = 0.0
v_0x = v_0 * math.cos(angulo)
# v_0x = v_0 * math.cos(angulo) * math.pow(math.e, -k * tempo)
v_0y = v_0 * math.sin(angulo)
# v_0y = - g / k + math.pow(math.e, - k * tempo) / k * (g + k * v_0 * math.sin(angulo))

dt = 0.1

while tempo <= 3.6:
    # x_t = x_0 + v_0x * tempo
    # x_t = x_0 + (v_0x / k) * (1 - math.pow(math.e, -k * tempo))
    # y_t = y_0 + v_0y * tempo - 0.5 * g * tempo ** 2
    # y_t = y_0 + (- g / k) * tempo + ((g + k * v_0y) / (k ** 2)) * (1 - math.pow(math.e, - k * tempo))

    x_t = v_0x / k * (1 - math.pow(math.e, - k * tempo))
    y_t = 1 / k * (v_0y + g / k) * (1 - math.pow(math.e, -k * tempo)) - g / k * tempo
    
    lista_tempo.append(tempo)
    lista_eixo_x.append(x_t)
    lista_eixo_y.append(y_t)
    
    tempo += dt

plt.xlabel('$Distância$ ($m$)')
plt.ylabel('$Altura$ (m)')
plt.plot(lista_eixo_x, lista_eixo_y)
plt.title('Movimento Projétil')
```
