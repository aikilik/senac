```python
import matplotlib.pyplot as plt
import math

lista_tempo = []
lista_eixo_x = []
lista_eixo_y = []

x_0 = 0.0
y_0= 0.0
g = 9.8
v_0 = 18.0
angulo = 80 * math.pi / 180

# dados para o fenomeno do arrasto
m = 5
b = 0.47

densidade_ar = 1.225 # densidade do ar
A = 0.8 # A = √°rea de se√ß√£o tranversal
C_x = 0.5 # coeficiente de arrasto aerodin√¢mico. Para uma bola o valor √© de 0.5

F_a = (densidade_ar * C_x * A * v_0 ** 2) / 2
 
tempo = 0.0
v_0x = v_0 * math.cos(angulo)
v_0y = v_0 * math.sin(angulo)


dt = 0.1

while tempo <= 3.6:
    x_t = x_0 + v_0x * tempo
    y_t = y_0 + v_0y * tempo - 0.5 * g * tempo ** 2
    
    lista_tempo.append(tempo)
    lista_eixo_x.append(x_t)
    lista_eixo_y.append(y_t)
    
    tempo += dt

plt.xlabel('$Dist√¢ncia$ ($m$)')
plt.ylabel('$Altura$ (m)')
plt.plot(lista_eixo_x, lista_eixo_y)
plt.title('Movimento Proj√©til')
```
