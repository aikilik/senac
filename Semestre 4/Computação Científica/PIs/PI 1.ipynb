Série de Taylor para a função $y = (-x)^{1/2}$


Derivadas:

$f¹(x) = -\frac{1}{2\sqrt{-x}}$

$f²(x) = -\frac{1}{4(-x)^{3/2}}$

$f³(x) = -\frac{3}{8(-x)^{5/2}}$

$f⁴(x) = -\frac{15}{16(-x)^{7/2}}$

$f⁵(x) = -\frac{105}{32(-x)^{9/2}}$

Série de Taylor em $a = -1$:

$1-\frac{x+1}{2}-\frac{1}{8}(x+1)²-\frac{1}{16}(x+1)³-\frac{5}{128}(x+1)⁴-\frac{7}{256}(x+1)⁵...$

Versão otimizada:

$\Sigma_{n=0}^{\infty}(-1+x)^n i\binom{1/2}{n}$

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(-10, 1, 100)

y = np.power(-x, 1/2)


fig, ax = plt.subplots()
fig.set_size_inches(20,10)
ax.plot(x, y, linewidth = 2.0, label="derivada 0")

ax.legend()
ax.grid()
```

```python
def s_taylor_sqrt_negativa(x, n = 5):
    derivada_0 = np.power(-x, 1/2)
    derivada_1 = derivada_0 / 2
    numerador_s = (np.power(-x, (2*n-1)/2)*np.power(x, n))
    denominador_s = 2*n*math.factorial(n)
    serie = numerador_s / denominador_s
    return derivada_0 + derivada_1 + serie

print(s_taylor_sqrt_negativa(-1))
```

```python
y0 = np.power(-x, 1/2)
y1 = s_taylor_sqrt_negativa(x)

fig, ax = plt.subplots()
fig.set_size_inches(20,10)
ax.plot(x, y0, linewidth = 2.0, label="derivada 0")
ax.plot(x, y1, linewidth = 2.0, label="Série de Taylor")

ax.legend()
ax.grid()
```
