```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12

# quantidades de produtos vendidos por mês
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
```

# Gráfico 1 - Total de produtos Vendidos por mês - Linha

```python
total_mes = np.array([])
for i in range(12):
    total_mes = np.append(total_mes, creme_facial[i] + limpeza_facial[i] + pasta_dentaria[i] + sabonete[i] + shampoo[i] + hidratante[i])

meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]
x = np.linspace(1, 12, 12)
y = total_mes

fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title("Total de produtos vendidos por mês")

ax.set_xticks(x)
ax.set_xticklabels(meses)
ax.set_ylabel("Soma das vendas por mês")

fig.set_size_inches(15, 10)

plt.show()
```

# Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha

```python
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]
x = np.linspace(1, 12, 12)

fig, ax = plt.subplots()

ax.set_title("Todos os produtos vendidos por mês")
ax.plot(x, creme_facial, label = "Creme Facial")
ax.plot(x, limpeza_facial, label = "Limpeza Facial")
ax.plot(x, pasta_dentaria, label = "Pasta Dentaria")
ax.plot(x, sabonete, label = "Sabonete")
ax.plot(x, shampoo, label = "Shampoo")
ax.plot(x, hidratante, label = "Hidratante")

ax.set_xticks(x)
ax.set_xticklabels(meses)
ax.set_ylabel("Quantidade de vendas")
ax.legend()

fig.set_size_inches(15, 10)

plt.show()
```

# Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras

```python
y1 = creme_facial
y2 = limpeza_facial

meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"]

fig, ax = plt.subplots()
largura_barra = 0.3
ax.bar(x - largura_barra / 2, y1, width = largura_barra, label = "Creme Facial")
ax.bar(x + largura_barra / 2, y2, width = largura_barra, label = "Limpeza Facial")

ax.set_xticks(x)
ax.set_xticklabels(meses)

ax.legend()

fig.set_size_inches(15, 10)

plt.show()
```

# Gráfico 5 - Pizza. % da quantidade produtos vendidos no ano em cada produto

```python
total_creme_facial = np.sum(creme_facial)
total_limpeza_facial = np.sum(limpeza_facial)
total_pasta_dentaria = np.sum(pasta_dentaria)
total_sabonete = np.sum(sabonete)
total_shampoo = np.sum(shampoo)
total_hidratante = np.sum(hidratante)

y = [total_creme_facial, total_limpeza_facial, total_pasta_dentaria, total_sabonete, total_shampoo, total_hidratante]
produtos = ["Creme facila", "Limpeza Facial", "Pasta Dental", "Sabonete", "Shampoo", "Hidratante"]

fig, ax = plt.subplots()

ax.pie(y, labels = produtos, shadow = True, autopct = "%1.1f%%")

plt.show()
```
