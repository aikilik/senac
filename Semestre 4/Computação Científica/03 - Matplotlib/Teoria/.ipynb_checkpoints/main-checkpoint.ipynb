# Gr√°fico de linha

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Hello World

```python
# senoide
# linspace divide o resultado de 0 a 10 em 100 pontos distintos
x = np.linspace(0, 10, 100) # sequ√™ncia de 0 a 10
y = 3 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()

ax.plot(x, y, "o") # plota (imprime) linhas, pontos etc

plt.show()
```

```python
# linspace divide o resultado de 0 a 10 em 100 pontos distintos
x = np.linspace(0, 10, 100) # sequ√™ncia de 0 a 10
y = 3 + 2 * np.sin(2 * x)

fig, axs = plt.subplots(1, 2) # 1 linha, 2 colunas

# primeiro gr√°fico
axs[0].plot(x, y)

# segundo gr√°fico
axs[1].plot(x, y, "o")

fig.set_size_inches(20,8)

plt.show()
```

Mesmo gr√°fico

```python
x = np.linspace(1, 10, 100)

fig, ax = plt.subplots()

fig.set_size_inches(10, 5.5)

def exponencial(n):
    return 2 ** n

# plotar as fun√ß√µes
ax.plot(x, np.log2(x), label = "logaritmico")
ax.plot(x, np.log2(x), label = "n * logaritmico")
ax.plot(x, x, label = "linear")
ax.plot(x, x ** 2, label = "quadr√°tico")
ax.plot(x, exponencial(x), label = "exponencial")

# plotar as labels
ax.set_title("Complexidade algoritmica")
ax.set_xlabel("N")
ax.set_ylabel("Complexidade")
ax.legend()

plt.show()
```

Estiliza√ß√£o das linhas

```python
x = np.linspace(0, 5, num = 100)
y = 4 + 2 * np.sin(2 * x)

fig, axs = plt.subplots(1, 3)

axs[0].plot(x, y, color = "green")
axs[0].set_title("Gr√°fico 1")

axs[1].plot(x, y, color = "blue", marker = "o", linewidth = 0)
axs[1].set_title("Gr√°fico 2")

axs[2].plot(x, y, color = "red", marker = "s", linewidth = 5, markersize = 10)
axs[2].set_title("Gr√°fico 3")

fig.set_size_inches(20, 8)

fig.savefig("grafico.png")

plt.show()
```

Histograma

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(20, 8) # tamanho largura x altura

# Gaussiana
# m√©dia, desvio
media, desvio = 50, 5
x = np.random.normal(media, desvio, 1000)
ax1.hist(x, edgecolor = "white")
ax1.set_title("Gaussiana")
ax1.grid()
ax1.text(40, 220, r"$\mu = 50, \sigma = 5$")


# Poisson
eventos = 3
x = np.random.poisson(eventos, 1000)
ax2.hist(x, edgecolor = "white")
ax2.set_title("Poisson")
ax2.grid()
ax2.text(4, 200, "eventos = 3")

plt.show()
```

Gr√°fico de barras

```python
np.random.seed(3)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

fig, ax = plt.subplots()

ax.bar(x, y, width = 1, edgecolor = "white")

ax.set(xlim = (0, 8), ylim = (0, 7))


plt.show()
```

```python
np.random.seed(3)
x = 0.5 + np.arange(8)

valores = ["T1", "T2", "T3", "T4", "T5", "T6", "T7", "T8"]
y1 = np.array([1, 5, 7, 5, 2, 6, 8, 9])
y2 = np.array([9, 4, 3, 6, 7, 5, 3, 6])

fig, ax = plt.subplots()
largura_barra = 0.3
ax.bar(x - largura_barra / 2, y1, width = largura_barra, label = "Grupo 1")
ax.bar(x + largura_barra / 2, y2, width = largura_barra, label = "Grupo 2")

ax.set_xticks(x)
ax.set_xticklabels(valores)

ax.legend()

plt.show()
```

Gr√°fico de pizza

```python
y = np.array([32,25,25,15])
textos = ["Bananas", "Ma√ß√£s", "Peras", "Uvas"]
explodir = [0.2,0,0,0]

fig, ax = plt.subplots()

ax.pie(y, labels = textos, shadow = True, explode = explodir, autopct = "%1.1f%%")

plt.show()
```
