### Parte 1

Série de Taylor para $f(x) = ln(x)$

Derivadas de $f(x) = ln(x)$ onde $x = 1$:

$f(x) = ln(x) \to f(1) = 0$

$f¹(x) = \frac{1}{x} \to f^1(1) = 1$

$f^2(x) = -\frac{1}{x^2} \to f²(1) = -1$

$f³(x) = \frac{2}{x³} \to f³(1) = 2$

$f⁴(x) = -\frac{(2.3)}{x⁴} \to f⁴(1) = -6$

$f⁵(x) = \frac{(2.3.4)}{x⁵} \to f⁵(1) = 24$

Logo, computacionalmente:

$ln(x) = (x-1) - \frac{1}{2}(x-1)²+\frac{1}{3}(x-1)³-\frac{1}{4}(x-1)⁴+\frac{1}{5}(x-1)⁵...$

Generalização:

$T(x) = \sum_{k=1}^{\infty} \frac{(-1)^{k+1}}{k}(x-1)^k$

### Parte 2

```python
import matplotlib.pyplot as plt
import numpy as np
import math

```

```python
def log_natural_otimizado(x, k_max = 5):
    log_natural = 0
    k = 1
    for k in range(1, k_max):
        numerador = np.power(-1, k + 1) * np.power(x-1, k)
        denominador = k
        log_natural += numerador / denominador
    return log_natural

x = np.linspace(0.4, 1.6, 100)

y1 = np.log(x)
y2 = log_natural_otimizado(x)

fig, ax = plt.subplots()
fig.set_size_inches(20,8)
ax.plot(x, y1, linewidth = 2.0, label="numpy")
ax.plot(x, y2, linewidth = 2.0, label="otimizado")
ax.legend()
ax.grid()
```

Para valores de $0.4 < x < 1.6$ a Série de Taylor é correta com $k = 5$.
